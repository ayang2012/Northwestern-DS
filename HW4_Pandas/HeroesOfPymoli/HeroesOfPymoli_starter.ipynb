```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#turns floats to currency
def to_currency(df, list_cols):
    for col in list_cols:
        df[col] = round(df[col], 2)
        df[col] = "$" + df[col].astype(str)
    
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data['binned'] = pd.cut(purchase_data['Age'], bins, labels=group_names)
df_binned = purchase_data.groupby('binned').nunique()

dem_table = pd.concat([df_binned['SN'], round(df_binned['SN']/df_binned['SN'].sum()*100,2)], axis=1)
dem_table.columns = ['Total Count', 'Percentage of Players']
dem_table['Percentage of Players'] = dem_table['Percentage of Players'].astype(str) + "%" 
dem_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df_binned = purchase_data.groupby('binned')
df_binned
#purchase_data.nunique()
purchasing_analysis = pd.DataFrame()
purchasing_analysis = pd.concat([ df_binned.nunique()['Purchase ID'], round(df_binned.mean()['Price'],2), df_binned.sum()['Price'], df_binned.sum()['Price']/df_binned.nunique()['SN']], axis=1)
purchasing_analysis.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']

for col in ['Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']:
    purchasing_analysis[col] = purchasing_analysis[col].map("${:,.2f}".format)

purchasing_analysis


```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_group = purchase_data.groupby('SN')
purchase_count = SN_group.nunique()['Purchase ID']
avg_purchase = SN_group.mean()['Price']
tot_purchase = SN_group.sum()['Price']

top_spenders = pd.concat([purchase_count, avg_purchase, tot_purchase], axis=1)
top_spenders.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']
top_spenders.sort_values(['Total Purchase Value'], ascending = False, inplace=True)

for col in ['Average Purchase Price', 'Total Purchase Value']:
    top_spenders[col] = top_spenders[col].map("${:.2f}".format)
top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_group = purchase_data.groupby('Item ID')

item_group['Item Name']
purchase_count = item_group.count()['Purchase ID']
item_group['Price']
tot_purchase = item_group.sum()['Price']

pop_items = pd.concat([purchase_count, tot_purchase], axis=1)
pop_items.columns = ['Purchase Count', 'Total Purchase Value']
pop_items.sort_values(['Total Purchase Value'], ascending = False, inplace=True)

og_df = purchase_data[['Item ID', 'Item Name', 'Price']]
og_df.set_index('Item ID', inplace=True)

# print(pop_items.head())
# print(og_df.head())

new_pop_items = pop_items.merge(og_df, left_index=True, right_index=True)
new_pop_items = new_pop_items.drop_duplicates(keep='first')
new_pop_items = new_pop_items.sort_values('Purchase Count', ascending = False)

new_pop_items = new_pop_items[['Item Name', 'Purchase Count', 'Price', 'Total Purchase Value']]
new_pop_items.rename(columns = {'Price':'Item Price'}, inplace=True)
for col in ['Total Purchase Value', 'Item Price']:
    new_pop_items[col] = new_pop_items[col].map("${:.2f}".format)
new_pop_items.head()
```
