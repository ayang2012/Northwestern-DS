```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse = pd.read_csv(mouse_drug_data_to_load)
df_clinical = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset

# Display the data table for preview

```

```python
df_mouse.head()
```

```python
df_clinical.head()
```

```python
df_all = pd.merge(df_mouse, df_clinical, on='Mouse ID')
df_all.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time_group = df_all.groupby(['Drug', 'Timepoint'])
drug_time_df = pd.DataFrame(drug_time_group.mean()['Tumor Volume (mm3)'])
# Convert to DataFrame

# Preview DataFrame
drug_time_mean = drug_time_df.reset_index()
drug_time_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
from scipy.stats import sem
drug_time_df = pd.DataFrame(drug_time_group['Tumor Volume (mm3)'].apply(lambda x: sem(x)))
# Convert to DataFrame

# Preview DataFrame
drug_time_sem = drug_time_df.reset_index()
drug_time_sem.head()
# Preview DataFrame

```

```python
# Minor Data Munging to Re-Format the Data Frames
df_mung = drug_time_mean.copy()
df_mung = df_mung.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
# Preview that Reformatting worked
df_mung.head()
```

```python
df_mung['Capomulin']
drug_time_sem[drug_time_sem['Drug'] == 'Capomulin']
```

```python
fig, ax = plt.subplots()
for col in df_mung.columns:
    ax.errorbar(df_mung.index, df_mung[col], drug_time_sem[drug_time_sem['Drug'] == col]['Tumor Volume (mm3)'], fmt="o", ls='-')
ax.legend()
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Reponse to Treatment')
```

```python
# Generate the Plot (with Error Bars)
# Setting up the plot
fig, ax = plt.subplots()
for col in ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']:
    ax.errorbar(df_mung.index, df_mung[col], drug_time_sem[drug_time_sem['Drug'] == col]['Tumor Volume (mm3)'], fmt="o", ls='-')
ax.legend()
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Reponse to Treatment')
plt.grid(True)
# Save the Figure

```

```python
# Show the Figure
plt.show()
```
